# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('Telco-Customer-Churn.csv')
```

```python
data.head()
```

# Comparison of churn count with respect to gender

```python
sns.countplot(x ='gender', hue='Churn', data=data)
```

# Female senior citizens

```python
fem_sc = data[(data['gender']=='Female') & (data['SeniorCitizen']==1)]
print('Number of female senior citizens =',fem_sc["gender"].count())
```

```python
sns.countplot(fem_sc['gender'])
plt.title('Count of female Senior Citizens')
```

# Comparison of tenure with Total Charges

```python
plt.figure(figsize=(6,6))
plt.scatter(data["tenure"],data["TotalCharges"],s=10,marker="*")
plt.xlabel("Tenure",fontsize=14)
plt.ylabel("Total Charges",fontsize=14)
plt.title("Relation between Tenure and Total charges",fontsize=18)
```

# contract preferred by the senior citizen

```python
plt.figure(figsize=(8,8))
sns.countplot(x=data['Contract'])
plt.title('Contract preference of Senior Citizen', fontsize=15)
```

# Finds on Payment Method

```python
plt.figure(figsize=(10,8))
sns.countplot(x=data['PaymentMethod'])
plt.title('Prefered payment method',fontsize=15)
```

```python
plt.figure(figsize=(10,8))
sns.countplot(data["PaymentMethod"],hue=data["Contract"])
plt.title("Contract wise Payment methods preferrence")
```

```python
plt.figure(figsize=(10,8))
sns.countplot(data["PaymentMethod"],hue=data["gender"])
plt.title("Gender wise Payment methods preferrence")
```
